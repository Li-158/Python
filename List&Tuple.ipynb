```python
# Python 特殊寫法，[運算示 for 項目 in 可替代物 (if 條件句)], 想成 for迴圈 在前面放上要加入的內容
number_list=[number for number in range(1,6)]
number_list
number_list=[number for number in range(1,6) if number%2==1]
number_list
```

```python
rows=range(1,4)
cols=range(1,3)
for row in rows:
    for col in cols:
        print(row, col)
#另一種快速的寫法
rows=range(1,4)
cols=range(1,3)
cells=[(row,col) for row in rows for col in cols]
for cell in cells:
    print(cell)
```
