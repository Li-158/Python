```python
#寫入檔案
#前面為檔案名稱, 後面為類型(mode) 第一個表示操作(w 代表寫入，如果檔案不存在，建立他) 後面字母表示類型(t 表文字, b表示二進位)
fout= open('oops.txt','wt')
fout.close()
```

```python
#可用print 寫入文字檔
fout = open('oops.txt', 'wt')
print("Oops, I created a file.", file= fout)
fout.close()

#也可以用 write 寫入文字檔, 但格式不一樣

poem ='''To see a world in a grain of sand,
And a heaven in a wild flower,
Hold infinity in the palm of your hand,
And eternity in an hour.
'''
fout= open('relativity.txt', 'wt')
fout.write(poem)
fout.close()
```

```python
#read(), readline(), readlines()來讀取文字檔
fin = open('relativity.txt','rt')
poem = fin.read()
fin.close()
len(poem)
```

```python
#提供最大字元來限制read()的數量
poem=''
fin = open('relativity.txt', 'rt')
chunk = 100
while True:
    fragment = fin.read(chunk)
    if not fragment:
        break
    poem += fragment

fin.close()
len(poem)
```

```python
#利用readline() 來一次讀取一行 
poem =''
fin = open('relativity.txt','rt')
while True:
    line = fin.readline()
    if not line:
        break
    poem += line
fin.close()
```
