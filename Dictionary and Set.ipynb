```python
#用[鍵]加入或改變項目, 鍵是唯一的 但值可以改變
pythons=dict(Chapman='Graham', Cleese='John',Idle='Eric',
            Jones='Terry',Palin='Michael')
pythons['Gillam']='Terry'
other=dict(Marx='Groucho',Howard='Moe')
pythons.update(other) # update 結合
pythons
del pythons['Howard'] # 
pythons
```

```python
# keys 找出所有的鍵, values 找出所有的值
pythons['Cleese']
pythons.get('Idlee','Not a Python')
pythons.keys()
pythons.values()
```

```python
signals=dict(green='go',yellow='go faster',red='smile for the camera')
save_signals=signals
signals['blue']='confuse everyone'
save_signals # save_signal 應為原本的signal 但這樣的方式會被改變
#So
signals=dict(green='go',yellow='go faster',red='smile for the camera')
save_signals=signals.copy()
signals['blue']='confuse everyone'
save_signals
```
