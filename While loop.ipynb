```python
while True:
    value = input("Integer, please[q to quit]:")
    if value =='q':
        break
    number =int(value)
    if number % 2 ==0:
        continue
    print(number,"squared is", number*number)
```

```python
number=[1,3,5]
position=0
while position < len(number):
    number=number[position]
    if number %2 ==0:
        print('Found even number ',number)
        break
        position+=1
    else: break
print("No even number found")
```
