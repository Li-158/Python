```python
import time 
# strftime() 將物件轉換成字串 
fmt = "It's %A, %B %d, %Y, local time %I:%M:%S%p" #在time模組裡面有固定的輸出說明 %Y=Year, %m= month, %B = 月名稱(Feb,July), %A為星期
t = time.localtime()
t   #time.struct_time(tm_year=2022, tm_mon=2, tm_mday=9, tm_hour=11, tm_min=12, tm_sec=8, tm_wday=2, tm_yday=40, tm_isdst=0)
time.strftime(fmt, t) #fmt 為前面設定的格式, 後面則為 剛剛定義的時間
#"It's Wednesday, February 09, 2022, local time 11:12:38AM"
```
