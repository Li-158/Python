```python
def my_range(first=0, last=10, step=1):
    number=first
    while number<last:
        yield number
        number += step
```

```python
#decorator 裝飾器，接收一個函示之後再回傳另一個函式
#用 decorator_it() 定義一個裝飾器，他會做出： 印出函數名稱和值 2.用引數來執行函式 3. 印出結果 4.回傳修改後的函示以供使用
def document_it(func):
    def new_function(*args, **kwargs):
        print('Running function:', func.__name__)
        print('Positional arguments:',args)
        print('Keyword arguments:',kwargs)
        result = func(*args, **kwargs)
        print('Result:',result)
        return result
    return new_function
def add_ints(a,b):
    return a+b
cooler_add_ints = document_it(add_ints)
cooler_add_ints(3,5)
```
