```python
color="mauve"
if color=="red":
    print("It's a tomato")
elif color == "green":
    print("It's a green pepper")
elif color =="bee purple":
    print("I don't know what it is, but only bees can see it")
else:
    print("I've never heard of the color",color)
    
# 將 if 需要判定的值和其他程式語言有相同的概念
```

```python
letters = "abcdefghijklmnopqrstuvwsxyz"
letters[10:]
letters[:10]
letters[5:10]
letters[::7]

# slice概念 = [start :] 在start後面的數值全部顯現
# [:end] 在end以前的數值顯像
# [start : end] 顯像兩個之間的值
# [::value] 從1開始間隔多少
#[start: end : value] 從start到end以value開始呈現
```
